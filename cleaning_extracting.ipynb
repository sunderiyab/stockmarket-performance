```python

import mplcursors  # Import the mplcursors library

# Fetch historical data for NKLA
ticker = "NKLA"
start_date = "2013-08-18"  
end_date = "2023-08-18"    

# Fetch the historical data
nkla_data = yf.download(ticker, start=start_date, end=end_date)

# Convert the downloaded data to a DataFrame
df = pd.DataFrame(nkla_data)

# Plot the trading volume over time with tooltips
fig, ax = plt.subplots(figsize=(10, 6))
line = ax.plot(df.index, df['Volume'], label='Trading Volume', color='green')[0]
ax.set_xlabel('Date')
ax.set_ylabel('Trading Volume')
ax.set_title('Trading Volume Over Time')

# Add tooltips using mplcursors
cursor = mplcursors.cursor(line, hover=True)
cursor.connect("add", lambda sel: sel.annotation.set_text(f"Volume: {sel.artist.get_ydata()[sel.index]:,.0f}\nDate: {df.index[sel.index].date()}"))

plt.legend()
plt.grid(True)
plt.show()
```

```python
#How volatile is the stock's price?
#Calculate Returns:
#Calculate the daily returns of the instrument's prices. Returns are calculated as the percentage change in price from one day to the next.

#calculate Volatility:
#Calculate the standard deviation of the daily returns. This standard deviation provides a measure of the instrument's volatility
 

import numpy as np

# Replace with the stock symbol you're interested in
symbol = "NKLA"

# Fetch historical data
stock = yf.Ticker(symbol)
historical_data = stock.history(period="1y")

# Calculate daily returns
historical_data["Daily_Return"] = historical_data["Close"].pct_change()

# Calculate volatility (standard deviation of returns)
volatility = np.std(historical_data["Daily_Return"])

print(f"Volatility for {symbol} last 1y: {volatility:.4f}")



```

# What is the short interest in the stock?

```python
import requests
from bs4 import BeautifulSoup

# URL of the page
url = 'https://www.nyse.com/quote/XNGS:NKLA/OPTIONS'

# Send a GET request to the URL
response = requests.get(url)
html = response.text

# Parse the HTML using BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')

# Find the element containing the short interest data
short_interest_element = soup.select_one('.options-header-cell-symbol')

if short_interest_element:
    short_interest_text = short_interest_element.get_text()
    print("Short Interest Calls:", short_interest_text)
else:
    print("Short interest element not found")

```
